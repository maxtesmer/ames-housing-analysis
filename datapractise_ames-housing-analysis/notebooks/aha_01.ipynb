```python
import pandas as pd
```

```python
# CSV 
df = pd.read_csv("../data/AmesHousing.csv")
df.head()
```

```python
df.describe().T
```

```python
correlation_matrix = df.corr(numeric_only=True) 
correlations = correlation_matrix["SalePrice"].sort_values(ascending=False)
correlations.head(10)

```

##  Ziel der Analyse (Tag 2)

Heute beginne ich mit der explorativen Datenanalyse (EDA) des Ames Housing Datensatzes.  
Der Fokus liegt auf der Frage:

> **Welche numerischen Variablen hängen am stärksten mit dem Verkaufspreis (`SalePrice`) zusammen?**

Um diese Frage zu beantworten:
- habe ich zuerst mit `df.describe()` die Verteilung der wichtigsten numerischen Spalten geprüft
- anschließend habe ich mit `df.corr()` berechnet, **welche numerischen Merkmale mit `SalePrice` korrelieren**

- Die Hausqualität (`OverallQual`) hat die stärkste Korrelation mit dem Preis (r ≈ 0.80)
- Wohnfläche (`GrLivArea`) und Garage (`Garage Cars`, `Garage Area`) sind ebenfalls starke Prädiktoren
- Das Baujahr hat auch eine relevante, aber schwächere Korrelation (r ≈ 0.55)

Diese Erkenntnisse helfen mir später bei der Auswahl relevanter Merkmale für Modelle und Visualisierungen.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#top zehn Einflussfaktoren auf 'SalesPrice' 
top_correlation_features = correlations.index[0:11]#0 = SalePrice

#Heatmap 
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix.loc[top_correlation_features, top_correlation_features], annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Top 10 Features with SalePrice")
plt.show() 
```

Die Heatmap zeigt die Korrelation zwischen den 10 wichtigsten numerischen Merkmalen **inkl. `SalePrice`**.

**Zentrale Erkenntnisse:**
- `OverallQual`, `GrLivArea` und `Garage Cars` zeigen die **höchste Korrelation mit dem Hauspreis**
- `Garage Area`, `Total Bsmt SF` und `1st Flr SF` sind ebenfalls stark positiv korreliert
- Viele dieser Merkmale korrelieren **untereinander ebenfalls stark**, z. B. `Garage Area` & `Garage Cars`, oder `Total Bsmt SF` & `1st Flr SF` → mögliches Zeichen für **Redundanz**

**Nächster Schritt:**  
Wir analysieren die **Korrelation zu `SalePrice` nochmal separat**, um die Werte isoliert darzustellen.

```python
plt.figure(figsize=(10, 1))
sns.heatmap(correlations.head(11).to_frame().T, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Korrelation mit dem Verkaufspreis")
plt.show() 
```

Die Heatmap zeigt, dass der Hauspreis (`SalePrice`) am stärksten mit folgenden Variablen korreliert:
- `OverallQual`: Qualität des Hauses
- `GrLivArea`: Wohnfläche
- `Garage Cars` & `Garage Area`: Größe und Anzahl der Garagenplätze
- `Total Bsmt SF`: Gesamtfläche des Kellers

Diese Variablen sind **gute Kandidaten** für weitere Analyse oder erste Modellbildung.
