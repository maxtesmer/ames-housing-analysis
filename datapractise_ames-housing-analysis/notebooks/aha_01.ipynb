**Gibt es ein optimales MaÃŸ an WohnflÃ¤che, GaragenplÃ¤tzen usw.,
bei dem man relativ zur Investition den grÃ¶ÃŸten Preissprung bekommt?**



**ROI = Return on Investment**  
â†’ Also: **Was bekomme ich zurÃ¼ck fÃ¼r das, was ich investiert habe?**

In unserem Fall heiÃŸt das:

> **Wie viel Verkaufspreis bringt mir jeder zusÃ¤tzliche QuadratfuÃŸ WohnflÃ¤che?**

Wir nennen das:  
**Preis pro QuadratfuÃŸ (`Preis_pro_sqft`)**

---

### Warum ist das sinnvoll?

- GroÃŸe HÃ¤user sind **nicht automatisch wirtschaftlicher**  
- Mehr FlÃ¤che = mehr Baukosten â†’ lohnt sich das auch?
- Irgendwann sagt der Markt: â€žSo viel FlÃ¤che brauch ich nichtâ€œ â†’ Preis stagniert oder sinkt

Mit `Preis_pro_sqft` analysieren wir **die Effizienz von WohnflÃ¤che**  
â†’ Also: Wie viel â€žWertâ€œ steckt in jedem QuadratfuÃŸ?

```python
df["preis_pro_sqft"]=df["SalePrice"]/df["Gr Liv Area"]

df["fleachenkategorie"]=pd.cut(df["Gr Liv Area"], bins=[0, 1000, 2000, 3000, 4000, 5000, 6000], labels=["<1000", "1000-2000", "2000-3000", "3000-4000", "4000-5000", ">5000"])
roi_means = df.groupby("fleachenkategorie")["preis_pro_sqft"].mean().sort_index(ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x=roi_means.index, y=roi_means.values)
plt.title("ROI pro Quadratmeter nach FlÃ¤chengrÃ¶ÃŸe")
plt.xlabel("FlÃ¤chengrÃ¶ÃŸe")
plt.ylabel("ROI pro Quadratmeter")
plt.xticks(rotation=45)
plt.show()
# Boxplot fÃ¼r die Preisverteilung       
plt.figure(figsize=(12, 6))
sns.boxplot(x="fleachenkategorie", y="preis_pro_sqft", data=df)
plt.title("Preis pro Quadratmeter nach FlÃ¤chengrÃ¶ÃŸe")
plt.xlabel("FlÃ¤chengrÃ¶ÃŸe")
plt.ylabel("Preis pro Quadratmeter")
plt.xticks(rotation=45)
plt.show()

```

## Analyse: Wann lohnt sich WohnflÃ¤che wirklich?

Basierend auf meiner Analyse in Bar- & Boxplot ergibt sich folgendes Bild:

- HÃ¤user mit **Ã¼ber 5000 sqft** WohnflÃ¤che sind **deutlich ineffizient**. Der ROI bricht extrem ein.
- Der **Sweetspot liegt zwischen 1000â€“3000 sqft**:
  - Hoher durchschnittlicher ROI
  - Geringe Streuung â†’ stabile Investitionsgrundlage
- Sehr kleine HÃ¤user (<1000 sqft) haben zwar ebenfalls hohen ROI, aber:
  - Oft stark limitiert in Ausstattung & Funktion
  - Ggf. weniger marktattraktiv â†’ kleinere Zielgruppe


### Fazit:
**Maximale Effizienz in der Mitte**. FlÃ¤chen zwischen 1000 und 3000 sqft liefern im Schnitt:
- **den hÃ¶chsten Return per FlÃ¤che**
- **die grÃ¶ÃŸte Sicherheit (geringe Streuung)**
- **die beste Balance zwischen Angebot & Nachfrage**

Diese Erkenntnis ist wichtig fÃ¼r:
- Feature Engineering im Modell
