##  Ziel der Analyse (Tag 2)

Heute beginne ich mit der explorativen Datenanalyse (EDA) des Ames Housing Datensatzes.  
Der Fokus liegt auf der Frage:

> **Welche numerischen Variablen hängen am stärksten mit dem Verkaufspreis (`SalePrice`) zusammen?**

Um diese Frage zu beantworten:
- habe ich zuerst mit `df.describe()` die Verteilung der wichtigsten numerischen Spalten geprüft
- anschließend habe ich mit `df.corr()` berechnet, **welche numerischen Merkmale mit `SalePrice` korrelieren**

- Die Hausqualität (`OverallQual`) hat die stärkste Korrelation mit dem Preis (r ≈ 0.80)
- Wohnfläche (`GrLivArea`) und Garage (`Garage Cars`, `Garage Area`) sind ebenfalls starke Prädiktoren
- Das Baujahr hat auch eine relevante, aber schwächere Korrelation (r ≈ 0.55)

Diese Erkenntnisse helfen mir später bei der Auswahl relevanter Merkmale für Modelle und Visualisierungen.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#top zehn Einflussfaktoren auf 'SalesPrice' 
top_correlation_features = correlations.index[0:11]#0 = SalePrice

#Heatmap 
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix.loc[top_correlation_features, top_correlation_features], annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Top 10 Features with SalePrice")
plt.show() 
```

Die Heatmap zeigt die Korrelation zwischen den 10 wichtigsten numerischen Merkmalen **inkl. `SalePrice`**.

**Zentrale Erkenntnisse:**
- `OverallQual`, `GrLivArea` und `Garage Cars` zeigen die **höchste Korrelation mit dem Hauspreis**
- `Garage Area`, `Total Bsmt SF` und `1st Flr SF` sind ebenfalls stark positiv korreliert
- Viele dieser Merkmale korrelieren **untereinander ebenfalls stark**, z. B. `Garage Area` & `Garage Cars`, oder `Total Bsmt SF` & `1st Flr SF` → mögliches Zeichen für **Redundanz**

**Nächster Schritt:**  
Wir analysieren die **Korrelation zu `SalePrice` nochmal separat**, um die Werte isoliert darzustellen.

```python
plt.figure(figsize=(10, 1))
sns.heatmap(correlations.head(11).to_frame().T, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Korrelation mit dem Verkaufspreis")
plt.show() 
```

Die Heatmap zeigt, dass der Hauspreis (`SalePrice`) am stärksten mit folgenden Variablen korreliert:
- `OverallQual`: Qualität des Hauses
- `GrLivArea`: Wohnfläche
- `Garage Cars` & `Garage Area`: Größe und Anzahl der Garagenplätze

Diese Variablen sind **gute Kandidaten** für weitere Analyse oder erste Modellbildung.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="Gr Liv Area", y="SalePrice")
plt.title("Gr Liv Area vs SalePrice")
plt.xlabel("Gr Liv Area")
plt.ylabel("SalePrice")
plt.show()
```

**Gibt es ein optimales Maß an Wohnfläche, Garagenplätzen usw.,
bei dem man relativ zur Investition den größten Preissprung bekommt?**



**ROI = Return on Investment**  
→ Also: **Was bekomme ich zurück für das, was ich investiert habe?**

In unserem Fall heißt das:

> **Wie viel Verkaufspreis bringt mir jeder zusätzliche Quadratfuß Wohnfläche?**

Wir nennen das:  
**Preis pro Quadratfuß (`Preis_pro_sqft`)**

---

### Warum ist das sinnvoll?

- Große Häuser sind **nicht automatisch wirtschaftlicher**  
- Mehr Fläche = mehr Baukosten → lohnt sich das auch?
- Irgendwann sagt der Markt: „So viel Fläche brauch ich nicht“ → Preis stagniert oder sinkt

Mit `Preis_pro_sqft` analysieren wir **die Effizienz von Wohnfläche**  
→ Also: Wie viel „Wert“ steckt in jedem Quadratfuß?

```python
df["preis_pro_sqft"]=df["SalePrice"]/df["Gr Liv Area"]

df["fleachenkategorie"]=pd.cut(df["Gr Liv Area"], bins=[0, 1000, 2000, 3000, 4000, 5000, 6000], labels=["<1000", "1000-2000", "2000-3000", "3000-4000", "4000-5000", ">5000"])
roi_means = df.groupby("fleachenkategorie")["preis_pro_sqft"].mean().sort_index(ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x=roi_means.index, y=roi_means.values)
plt.title("ROI pro Quadratmeter nach Flächengröße")
plt.xlabel("Flächengröße")
plt.ylabel("ROI pro Quadratmeter")
plt.xticks(rotation=45)
plt.show()
# Boxplot für die Preisverteilung       
plt.figure(figsize=(12, 6))
sns.boxplot(x="fleachenkategorie", y="preis_pro_sqft", data=df)
plt.title("Preis pro Quadratmeter nach Flächengröße")
plt.xlabel("Flächengröße")
plt.ylabel("Preis pro Quadratmeter")
plt.xticks(rotation=45)
plt.show()

```

## Analyse: Wann lohnt sich Wohnfläche wirklich?

Basierend auf meiner Analyse in Bar- & Boxplot ergibt sich folgendes Bild:

- Häuser mit **über 5000 sqft** Wohnfläche sind **deutlich ineffizient**. Der ROI bricht extrem ein.
- Der **Sweetspot liegt zwischen 1000–3000 sqft**:
  - Hoher durchschnittlicher ROI
  - Geringe Streuung → stabile Investitionsgrundlage
- Sehr kleine Häuser (<1000 sqft) haben zwar ebenfalls hohen ROI, aber:
  - Oft stark limitiert in Ausstattung & Funktion
  - Ggf. weniger marktattraktiv → kleinere Zielgruppe


### Fazit:
**Maximale Effizienz in der Mitte**. Flächen zwischen 1000 und 3000 sqft liefern im Schnitt:
- **den höchsten Return per Fläche**
- **die größte Sicherheit (geringe Streuung)**
- **die beste Balance zwischen Angebot & Nachfrage**

Diese Erkenntnis ist wichtig für:
- Feature Engineering im Modell
- Investitionsstrategien
- Zielgruppenanalysen
